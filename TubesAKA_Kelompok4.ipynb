#Algoritma Selection Sort

Selection sort merupakan sebuah teknik pengurutan dengan cara mencari nilai tertinggi / terendah di dalam array kemudian menempatkan nilai tersebut di tempat semestinya.

image.png


## Source Code

```python
#Selection Sort Iterative
def selectionSortIterative(N):
    for i in range(len(N) - 1):
        min_idx = i
        for j in range(i + 1, len(N)):
            if N[j] < N[min_idx]:
                min_idx = j
        temp = N[min_idx]
        N[min_idx] = N[i]
        N[i] = temp
```

```python
nData = [1, 3, 0, 1, 2, 0, 4, 1, 8,0]
selectionSortIterative(nData)
print(nData)
```

## Analisis Kompleksitas

### Input Size

Selection Sort iterative memiliki input size **n-1**

### Basic Operation

### Number of times the basic operation

**1 kali** tiap iterasi

## Perhitungan Kompleksitas Algoritma

\begin{equation}
    T(n) = \sum _{i=0}^{n-1}\sum _{j=i+1}^n\:= \sum _{i=0}^{n-1}\:n-\left(i+1\right)+1\:= \sum _{i=0}^{n-1}\:n-i
\end{equation}
\begin{equation}
   T(n) = n\sum _{i=0}^{n-1}\:1\:-\:\sum _{i=0}^{n-1}\:i\:=\:n\sum _{i=0}^{n-1}\:1\:-\frac{\left(n-1\right)n}{2}
\end{equation}
\begin{equation}
   T(n) = n^2-\frac{\left(n-1\right)n}{2}\:=\:\frac{n\left(n+1\right)}{2}
\end{equation}
\begin{equation}
   T(n) = \frac{n^2+n}{2}\:=\:\frac{1}{2}n^2+n\:\in O(n^2)
\end{equation}

### Notasi Asimtotik

Pada algoritma selection sort iterative memiliki notasi asimtotik $ O(n^2) $

### Kelas Efisiensi

Selection sort iterative memiliki kelas efisiensi **Quadratic**

# Algoritma Insertion Sort

### Basic Operation

### Source Code

```python
#Insertion Sort iterative
def insertionSort(arr):
    for i in range(2, len(arr)):
  
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
```

```python
arr = [1, 3, 0, 1, 2, 0, 4, 2, 5, 1]
insertionSort(arr)
print ("Sorted array is:", arr)
```

## Analisis Kompleksitas

\begin{equation}
    T(n) = \sum _{i=2}^n\:\sum _{j=1}^i1\:=\:\sum _{i=2}^n\:\left(i-1\right)+1\:=\sum _{i=2}^n\:
\end{equation}
\begin{equation}
    =\left(\sum _{i=1}^ni\:\right)\:-1\:=\:\left(\frac{n\left(n+1\right)}{2}\right)-2
\end{equation}
\begin{equation}
    =\:\frac{n^2+2n-2}{2}\:=\:\frac{\left(n+2\right)\left(n-1\right)}{2}\:\in O\left(n^2\right)
\end{equation}

### Input Size

Insertion Sort iterative memiliki input size **n**

### Number of Times the Basic Operation

**1 kali** tiap iterasi

### Notasi Asimtotik

Pada algoritma insertion sort memiliki notasi asimtotik $ O(n^2) $

### Kelas Efesiensi

Insertion sort iterative memiliki kelas efisiensi **Quadratic**
